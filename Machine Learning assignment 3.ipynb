```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = .25,random_state =1) #data splitting
```

# Through RandomForest Algorithm:

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
r = RandomForestClassifier()
```

```python
r.fit(xtrain,ytrain)
```

```python
r.predict(xtest)
```

```python
r.score(xtest,ytest)
```

# Through Decision Tree Algorithm:

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dct = DecisionTreeClassifier()
```

```python
dct.fit(xtrain,ytrain)
```

```python
dct.score(xtest,ytest) #To measure accuracy
```
