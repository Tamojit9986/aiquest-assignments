# Ordinal Encoder()

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("shop data.csv")
df
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df["income"].unique()
```

```python
order = ["low","medium","high"]
oe = OrdinalEncoder(categories=[order])
```

```python
oe.fit(df[["income"]])
```

```python
final = pd.DataFrame(oe.transform(df[["income"]]))
```

```python
df[["income"]] = final
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
```
