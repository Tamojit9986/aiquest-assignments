# Decision Tree

```python
import pandas as pd
```

```python
df = pd.read_csv("shop data.csv")
df
```

```python
from sklearn import tree
```

```python
x = df.iloc[:, :-1]
x
```

```python
y = df.iloc[:, -1]
y
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
x = x.apply(le.fit_transform) #converting the string data into numeric type
```

```python
x
```

```python
classify_ = tree.DecisionTreeClassifier()
```

```python
classify_.fit(x,y)
```

```python
classify_.predict([[1,0,1,1]])
```
