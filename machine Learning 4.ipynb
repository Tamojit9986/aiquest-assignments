# Gradient Descent

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("car driving risk analysis.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
x = df["speed"]
y = df["risk"]
```

```python
n = int(len(x)) #no of rows
m = 0 #gradient
c = 0 #intercept
y_pred = m*x+c #y prediction formula
learning_rate = 0.001
iteration = 1000 #iterative process for Gradient Descent
```

```python
for i in range(iteration):
    derivative_m = (-2/n)*sum(x*(y-y_pred)) #formula to find derivatives of m
    derivative_c = (-2/n)*sum(y-y_pred) #formula to find the derivatives of c
    #Final step is updating the value of m and c
    m = m-learning_rate*derivative_m
    c = c-learning_rate*derivative_c
    print(m,c)
print("m value is", m)
print("c value is", c)
```
